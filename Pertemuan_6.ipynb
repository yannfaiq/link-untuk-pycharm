```python
def F(n):
    # Menghitung bilangan Fibonacci ke-n secara rekursif dengan menggunakan definisinya
    # Input: Sebuah bilangan bulat nonnegatif n
    # Output: Angka Fibonacci ke-n
    if n <= 1:
        return n
    else:
        return F(n - 1) + F(n - 2)
```

```python
n = int(input("Masukkan jumlah angka Fibonacci: "))
a, b = 0, 1
count = 0
while count < n:
    print(a, end=' ')
    a, b = b, a + b
    count += 1
```

```python
def greedy_coin_change(coins, amount):
    result = []
    coins.sort(reverse=True)
    for coin in coins:
        while amount >= coin:
            amount -= coin
            result.append(coin)
    return result

coins = [5, 4, 3, 1]
amount = 7
print("Kombinasi koin:", greedy_coin_change(coins, amount))
```

```python
import itertools

# Representasi kota dan jarak
def distance(A, B):
    # contoh jarak Euclidean
    return ((A[0] - B[0])**2 + (A[1] - B[1])**2) ** 0.5

def total_distance(tour):
    return sum(distance(tour[i], tour[i-1]) for i in range(len(tour)))

# Semua kemungkinan urutan tur
cities = [(0,0), (1,2), (3,1)]
tours = list(itertools.permutations(cities))
for t in tours:
    print(t, "Total jarak:", total_distance(t))
```

```python
import heapq

class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq

def huffman_encoding(chars, freqs):
    heap = [Node(c, f) for c, f in zip(chars, freqs)]
    heapq.heapify(heap)

    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        newNode = Node(None, left.freq + right.freq)
        newNode.left = left
        newNode.right = right
        heapq.heappush(heap, newNode)

    return heap[0]

def print_codes(node, code=''):
    if node:
        if node.char:
            print(f"{node.char}: {code}")
        print_codes(node.left, code + '0')
        print_codes(node.right, code + '1')

# Contoh data
chars = ['B', 'C', 'A', 'D']
freqs = [1, 6, 5, 3]

root = huffman_encoding(chars, freqs)
print_codes(root)
```
