```python
# SOAL 1
ef hitung_genap(arr):
    count = 0
    for x in arr:
        if x % 2 == 0:
            count += 1
    return count

data = [1, 2, 4, 7, 10]
print(hitung_genap(data))

#Penjelasan Aku menggunakan algoritma iterasi linear (linear scan):
#Loop melalui setiap elemen array. Periksa apakah elemen tersebut genap menggunakan x % 2 == 0. Jika genap, tambahkan counter. Kompleksitas waktu: O(n).
#Kasus khusus (edge cases) apa yang perlu diperhatikan?
# Array kosong → hasilnya 0. Elemen bukan bilangan (jika input tidak tervalidasi) → dapat menyebabkan error saat cek modulus. Bilangan negatif tetap bisa dicek dengan modulus (−4 tetap genap).
```

```python
# SOAL 2
def nilai_terbesar(arr):
    if not arr:
        return None
    maks = arr[0]
    for x in arr[1:]:
        if x > maks:
            maks = x
    return maks

data = [5, 2, 9, 1]
print(nilai_terbesar(data))

#Ambil elemen pertama sebagai nilai awal maksimum.

#Iterasi seluruh elemen.Setiap kali menemukan nilai lebih besar dari maksimum saat ini, perbarui maksimum. Setelah semua elemen diperiksa, kembalikan nilai maksimum.

#Apa yang harus dilakukan jika array kosong? Array kosong tidak memiliki nilai maksimum → harus: Mengembalikan None, atau Mengeluarkan error seperti ValueError("array kosong"). (Tergantung kebutuhan fungsi.)
```

```python
# SOAL 3
def balik_string(s):
    return s[::-1]

text = "halo"
print(balik_string(text))

#Logikanya adalah membaca karakter dari belakang ke depan, lalu menyusunnya kembali menjadi string baru: Bisa dengan loop menurun dari indeks terakhir. Atau dengan metode slicing s[::-1] yang secara otomatis membalik urutan karakter.
```

```python
# SOAL 4
def jumlah_elemen(arr):
    total = 0
    for x in arr:
        total += x
    return total

data = [3, 1, 2]
print(jumlah_elemen(data))

#Inisialisasi variabel total = 0. Iterasi setiap elemen dalam array. Tambahkan tiap elemen ke variabel total (total += x). Setelah loop selesai, kembalikan nilai total. Kompleksitas waktu: O(n).
```

```python
# SOAL 5
def cek_palindrome(s):
    return s == s[::-1]

print(cek_palindrome("level"))
print(cek_palindrome("hello"))

#Bandingkan string asli dengan string yang dibalik. Jika keduanya sama persis, berarti string itu palindrome. Contoh: "level" == "level" → palindrome.
```

```python
# SOAL 6
def hitung_vokal(s):
    vokal = "aiueoAIUEO"
    count = 0
    for ch in s:
        if ch in vokal:
            count += 1
    return count

text = "informatika"
print(hitung_vokal(text))

#Loop melalui setiap karakter dalam string. Periksa apakah karakter tersebut termasuk dalam daftar vokal: {a, i, u, e, o} (biasanya juga versi uppercase). Jika iya, tambahkan counter.
```

```python
# SOAL 7
def cari_elemen(arr, target):
    for x in arr:
        if x == target:
            return True
    return False

data = [4, 7, 1, 9]
print(cari_elemen(data, 7))

#Aku menggunakan linear search: Iterasi setiap elemen dalam array. Jika elemen sama dengan target, langsung kembalikan True. Jika sudah sampai akhir dan tidak ada yang cocok, kembalikan False. Kompleksitas waktu: O(n).
```
