```python
graph = {
    "Amin": ["Wasim", "Nick", "Mike"],
    "Wasim": ["Imran"],
    "Nick": [],
    "Mike": ["Faras"],
    "Imran": [],
    "Faras": []
}

def bfs(start):
    visited = []
    queue = [start]
    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])
    return visited

print(bfs("Amin"))
```

```python
graph = {
    "Amin": ["Wasim", "Nick", "Mike"],
    "Wasim": ["Imran"],
    "Nick": [],
    "Mike": ["Faras"],
    "Imran": [],
    "Faras": []
}

def bfs(start):
    visited = []
    queue = [start]
    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])
    return visited

def dfs(start):
    visited = []
    stack = [start]
    while stack:
        node = stack.pop()
        if node not in visited:
            visited.append(node)
            stack.extend(graph[node][::-1])
    return visited

print("BFS Amin:", bfs("Amin"))
print("DFS Amin:", dfs("Amin"))
print("BFS Wasim:", bfs("Wasim"))
print("DFS Wasim:", dfs("Wasim"))
print("BFS Faras:", bfs("Faras"))
print("DFS Faras:", dfs("Faras"))
```
