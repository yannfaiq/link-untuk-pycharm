```python
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.size())     # 3
print(q.dequeue())  # 1
print(q.isEmpty())  # False
```

```python
class Queue(object):
    def __init__(self):
        self.items = []
    def isEmpty(self):
        return self.items == []
    def enqueue(self, item):
        self.items.insert(0, item)
    def dequeue(self):
        return self.items.pop()
    def size(self):
        return len(self.items)
```
