```python
def polynomial_py(a, x):
    result = 0
    # for every n in 0..len(a)-1
    for n, a_n in enumerate(a):
        # compute x^n
        x_power_n = 1
        for i in range(n):
            x_power_n *= x
        # add a_n* x^n to the final result
        result += a_n * x_power_n
    return result
```

```python
a = [1, 2, 0, 3] # coefficients
x = 1.5
print(polynomial_py(a, x)) # 14.125
```

```python
def string_match(string, sub_str):
    # Brute force string matching
    for i in range(len(string)-len(sub_str)+1):
        index = i # index Point to the 1 Three characters to be compared
        for j in range(len(sub_str)):
            if string[index] == sub_str[j]:
                index += 1
            else:
                break
        if index-i == len(sub_str):
            return i
    return -1

if __name__ == "__main__":
    print(string_match("adbcbdc", "bdc"))
```

```python
import random
from itertools import permutations
from math import sqrt

# Define a simple City class
class acity:
    def __init__(self, x, y):
        self.real = x
        self.imag = y

    def __repr__(self):
        return f"acity({self.real}, {self.imag})"

def distance_tour (aTour):
    return sum(distance_points (aTour [i - 1], aTour[i]) for i in range(len(aTour)))

def distance_points (first, second):
    return sqrt((first.real - second.real)**2 + (first.imag - second.imag)**2)

def generate_cities (number_of_cities):
    seed=111;width=500;height=300
    random.seed(seed)
    return frozenset(acity(random.randint(1, width), random.randint(1,height)) for c in range(number_of_cities))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
def visualize_tour (tour, style='bo-'):
    if len(tour) > 1000: plt.figure(figsize=(15, 10))
    start = tour [0:1]
    visualize_segment (tour + start, style)
    visualize_segment (start, 'rD')
def visualize_segment (segment, style='bo-'):
    plt.plot([X(c) for c in segment], [Y(c) for c in segment], style, clip_on=False)
    plt.axis('scaled')
    plt.axis('off')
def X(city): "X axis"; return city.real
def Y(city): "Y axis"; return city.imag
```

```python
from time import process_time
from collections import Counter
def tsp(algorithm, cities):
    to = process_time()
    tour = algorithm(cities)
    t1 = process_time()
    assert Counter(tour) == Counter(cities)
    visualize_tour(tour)
    print("{}:{} cities => tour length {:.0f} (in {:.3f} sec)".format(name(algorithm), len(tour), distance_tour (tour), t1-to))
def name(algorithm): return algorithm.__name__.replace('_tsp','')
```

```python
def brute_force(cities):
    "Generate all possible tours of the cities and choose the shortest tour."
    return shortest_tour (alltours (cities))

def shortest_tour (tours):
    return min (tours, key=distance_tour)

def alltours(cities):
    "Return a list of all possible tours of the cities."
    return list(permutations(cities))
```

```python
#1
# Algoritma Brute Force: Sequential Search
# Mencari huruf 'g' dari deret a sampai h

def sequential_search(data, target):
    n = len(data)
    k = 0
    while k < n and data[k] != target:
        k += 1

    if k < n and data[k] == target:
        idx = k + 1  # posisi 1-based
    else:
        idx = 0      # tidak ditemukan
    return idx


# Deret huruf dari a sampai h
deret = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
target = 'g'

# Jalankan pencarian
hasil = sequential_search(deret, target)

# Tampilkan hasil
if hasil != 0:
    print(f"Huruf '{target}' ditemukan pada posisi ke-{hasil}.")
else:
    print(f"Huruf '{target}' tidak ditemukan.")
```

```python
#2
# Prosedur Pencarian Beruntun (Sequential Search)

def pencarian_beruntun(arr, x):
    """
    Mencari nilai x di dalam list arr.
    Jika ditemukan, mengembalikan indeks (1-based seperti pada algoritma di gambar).
    Jika tidak ditemukan, mengembalikan 0.
    """
    n = len(arr)
    k = 0  # karena indeks Python mulai dari 0

    # loop mencari sampai akhir atau sampai elemen ditemukan
    while k < n and arr[k] != x:
        k += 1

    # cek hasil
    if k < n and arr[k] == x:
        idx = k + 1  # ubah jadi 1-based index agar sesuai algoritma
    else:
        idx = 0      # tidak ditemukan

    return idx


# Contoh penggunaan:
data = [5, 3, 8, 6, 7, 2]
x = 6
hasil = pencarian_beruntun(data, x)

if hasil != 0:
    print(f"Elemen {x} ditemukan pada posisi ke-{hasil}.")
else:
    print(f"Elemen {x} tidak ditemukan.")
```

```python
#3
# Brute-force string matching untuk mencari pattern "56"

def brute_force_search_all(text, pattern):
    n = len(text)
    m = len(pattern)
    occurrences = []
    for i in range(n - m + 1):
        j = 0
        while j < m and text[i + j] == pattern[j]:
            j += 1
        if j == m:
            occurrences.append(i)  # simpan indeks 0-based
    return occurrences

if __name__ == "__main__":
    text = "123456789"   # contoh teks
    pattern = "56"
    occ = brute_force_search_all(text, pattern)

    if occ:
        print(f'Pattern "{pattern}" ditemukan.')
        print("Indeks (0-based):", occ)
        print("Posisi (1-based):", [i+1 for i in occ])
    else:
        print(f'Pattern "{pattern}" tidak ditemukan dalam teks.')
```

```python
#4
# tsp_bruteforce_vs_greedy.py
import math
import time
import random
import itertools
from typing import List, Tuple

Point = Tuple[float, float]

def factorial_permutations_for_tsp(n: int) -> int:
    """Jumlah rute yang harus dicek brute-force jika kita fix satu kota sebagai start: (n-1)!"""
    if n < 1:
        return 0
    return math.factorial(n - 1)

def human_time(seconds: float) -> str:
    yrs = seconds / (3600 * 24 * 365)
    if yrs >= 1:
        return f"{yrs:.2e} years ({seconds:.2f} seconds)"
    days = seconds / (3600 * 24)
    if days >= 1:
        return f"{days:.2f} days ({seconds:.2f} seconds)"
    hrs = seconds / 3600
    if hrs >= 1:
        return f"{hrs:.2f} hours ({seconds:.2f} seconds)"
    mins = seconds / 60
    if mins >= 1:
        return f"{mins:.2f} minutes ({seconds:.2f} seconds)"
    return f"{seconds:.2f} seconds"

def estimate_bruteforce_time(n: int, attempts_per_second_list=[1e3, 1e6, 1e9, 1e12]):
    perms = factorial_permutations_for_tsp(n)
    print(f"TSP n={n}: jumlah rute (fix start) = (n-1)! = {perms:,}")
    for s in attempts_per_second_list:
        sec = perms / s
        print(f"  Jika mencoba {int(s):,} rute/detik -> {human_time(sec)}")

# --- Greedy (Nearest Neighbor) TSP ---
def euclidean(a: Point, b: Point) -> float:
    return ((a[0]-b[0])**2 + (a[1]-b[1])**2)**0.5

def greedy_nearest_neighbor(points: List[Point], start_index: int = 0) -> Tuple[List[int], float]:
    n = len(points)
    if n == 0:
        return [], 0.0
    visited = [False] * n
    tour = [start_index]
    visited[start_index] = True
    total = 0.0
    current = start_index
    for _ in range(n-1):
        # find nearest unvisited
        next_idx = None
        next_dist = float('inf')
        for j in range(n):
            if not visited[j]:
                d = euclidean(points[current], points[j])
                if d < next_dist:
                    next_dist = d
                    next_idx = j
        tour.append(next_idx)
        visited[next_idx] = True
        total += next_dist
        current = next_idx
    # kembali ke start
    total += euclidean(points[current], points[start_index])
    return tour, total

# --- Brute-force exact TSP (gunakan HATI-HATI, hanya untuk n kecil) ---
def tsp_bruteforce_exact(points: List[Point]) -> Tuple[List[int], float]:
    n = len(points)
    if n <= 1:
        return list(range(n)), 0.0
    if n > 10:
        raise ValueError("tsp_bruteforce_exact: n terlalu besar; jalankan hanya untuk n <= 10")
    def tour_distance(order):
        d = 0.0
        for i in range(len(order)):
            a = points[order[i]]
            b = points[order[(i+1) % n]]
            d += euclidean(a, b)
        return d
    best = None
    best_d = float('inf')
    # fix kota 0 to remove rotational symmetry
    for perm in itertools.permutations(range(1, n)):
        order = (0,) + perm
        d = tour_distance(order)
        if d < best_d:
            best_d = d
            best = list(order)
    return best, best_d

# --- Demo / main ---
if __name__ == "__main__":
    random.seed(42)

    # 1) Estimasi waktu brute-force untuk n=20
    n = 20
    print("=== Estimasi Brute-Force untuk TSP n=20 ===")
    estimate_bruteforce_time(n, attempts_per_second_list=[1e3, 1e6, 1e9, 1e12])
    print()

    # 2) Bandingkan greedy runtime pada instance n=20
    n = 20
    points_20 = [(random.random()*100, random.random()*100) for _ in range(n)]
    print("=== Menjalankan Greedy (Nearest Neighbor) pada 20 titik acak ===")
    t0 = time.time()
    tour_greedy, dist_greedy = greedy_nearest_neighbor(points_20, start_index=0)
    t1 = time.time()
    print(f"Greedy selesai dalam {t1-t0:.6f} detik. Jarak tour: {dist_greedy:.4f}")
    print(f"Tour (indeks): {tour_greedy}")
    print()

    # 3) Contoh: jangan jalankan brute-force exact untuk n=20 (demo n kecil instead)
    print("=== Contoh penggunaan brute-force exact untuk n kecil (n=8) ===")
    n_small = 8
    points_small = [(random.random()*100, random.random()*100) for _ in range(n_small)]
    t0 = time.time()
    best_tour, best_dist = tsp_bruteforce_exact(points_small)
    t1 = time.time()
    print(f"Brute-force exact untuk n={n_small} selesai dalam {t1-t0:.4f} detik. Best distance: {best_dist:.4f}")
    print(f"Best tour: {best_tour}")
    print()
    print("Catatan: untuk n=20, brute-force exact TSP **tidak praktis** (lihat estimasi di atas). Greedy jauh lebih cepat walau tidak selalu optimal.")
```

```python
#5
# Fungsi untuk mencari nilai terkecil
def nilai_minimal(deret_bilangan):
    nilai_terkecil = deret_bilangan[0]

    for nilai in deret_bilangan:
        if nilai < nilai_terkecil:
            nilai_terkecil = nilai

    return nilai_terkecil


# Contoh data
a = [3, 20, 100, -35, 50]

print(a)
print('Nilai terkecil:', nilai_minimal(a))
```
