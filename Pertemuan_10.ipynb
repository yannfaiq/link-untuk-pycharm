```python
import pulp
from pulp import LpProblem, LpMinimize, LpVariable

# 4x + 3y = 34

# Membuat objek masalah optimasi
model = LpProblem(name="Pemecahan_Persamaan", sense=LpMinimize)

# Membuat variabel x dan y
x = LpVariable(name="x")
y = LpVariable(name="y")

# Menambahkan batasan
model += 4 * x + 3 * y == 34

# Menyelesaikan masalah
model.solve()

# Menampilkan hasil
print("Status:", model.status)
print("Nilai x =", x.varValue)
print("Nilai y =", y.varValue)
```

```python
import pulp

model = pulp.LpProblem("Harga_Pensil_dan_Buku", pulp.LpMinimize)

P = pulp.LpVariable("P", lowBound=0)
B = pulp.LpVariable("B", lowBound=0)

model += 0
model += 3 * P + 4 * B == 11000
model += P + 7 * B == 15000

model.solve()

print("Status:", pulp.LpStatus[model.status])
print("Harga satu pensil (P) =", pulp.value(P))
print("Harga satu buku (B) =", pulp.value(B))
print("Harga dua pensil =", 2 * pulp.value(P))
print("Harga enam buku =", 6 * pulp.value(B))
print("Total =", 2 * pulp.value(P) + 6 * pulp.value(B))
```

```python
from scipy.optimize import linprog

c = [-5, -7]
A = [
    [1, 0],
    [2, 3],
    [1, 1]
]
b = [16, 19, 8]

x1_bounds = (0, None)
x2_bounds = (0, None)

result = linprog(c, A_ub=A, b_ub=b, bounds=[x1_bounds, x2_bounds], method="highs")

print("Optimal Solution:")
print("x1 =", result.x[0])
print("x2 =", result.x[1])
print("Max Z =", -result.fun)
```

```python
#LATIHAN NO 1 DAN 2
import numpy as np
from scipy import linalg

# Sistem persamaan:
# 4x + 12y = 28
# 2x +  y  = 21

# Matriks koefisien (A)
A = np.array([[4, 12],
              [2,  1]])

# Matriks hasil (b)
b = np.array([28, 21])

# Menyelesaikan sistem persamaan Ax = b
x = linalg.solve(A, b)

# Menampilkan hasil
print("Nilai x dan y adalah:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
```

```python
#LATIHAN NO 3
import numpy as np
from scipy import linalg

# Sistem persamaan:
# x + y = 16
# 3x + 4y = 55

# Matriks koefisien (A)
A = np.array([[1, 1],
              [3, 4]])

# Matriks hasil (b)
b = np.array([16, 55])

# Menyelesaikan sistem persamaan Ax = b
x = linalg.solve(A, b)

# Membulatkan hasil ke 2 angka di belakang koma
x_rounded = np.round(x, 2)

# Menampilkan hasil
print("Hasil perhitungan (dibulatkan):")
print(f"Lisa bekerja selama {x_rounded[0]} jam")
print(f"Muri bekerja selama {x_rounded[1]} jam")
```
